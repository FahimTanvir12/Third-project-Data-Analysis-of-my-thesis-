```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import matplotlib as matplt 
```

```python
df0=pd.read_csv("./friday.csv")
df0.head()

df0["Temperature"]=df0["Unnamed: 0"].str[13:17]
df0["Humidity"]=df0["Unnamed: 0"].str[33:38]
df0.head()
```

```python
#Removing column Unnamed 0 and unnamed 2


df=df0.drop(["Unnamed: 0","Unnamed: 2"],axis=1)
df.to_csv("Series_of_data_set.csv")
df.head()
```

```python
#DESCRIBING DATA 

df.count()
df.describe()
```

```python
df.shape
df.isnull().sum()
```

```python
#Removing null rows 
df.isnull().sum()
df=df.dropna()
df.shape
```

```python

#DROPPING DUPLICATES 

df.duplicated().sum()
df.drop_duplicates()
df.shape
df.head()
```

```python
df.dropna(how="any")
df.replace(r'^\s*$',None, regex=True, inplace=True)
df.dropna()
```

```python
df.dropna()
#df["Temperature_filtered"]=df["Temperature"]!=None.drop()
df=df[df["Temperature"]!=None]
df=df[df["Humidity"]!=None]
df=df[df['Temperature'].str[:] !=None].dropna()
df.dropna()
```

```python
df['Humidity'] = pd.to_numeric(df['Humidity'], errors='coerce')

def humidity(filter):
    if filter == "nan%":
        return None
    else:
        return filter
df['Humidity']=df['Humidity'].apply(lambda x:humidity(x))
df.dropna()
df.to_csv("lets_see.csv")

df.head()
```

```python
df['Humidity'] = df['Humidity'].replace('nan%', pd.NA)
df['Humidity'] = pd.to_numeric(df['Humidity'], errors='coerce')
df = df.dropna(subset=['Humidity'])
df = df.dropna()
df.shape
```

```python
creating_a_timerange=pd.date_range(start="11:00 AM",end="3:00 PM ",periods=380).strftime('%I:%M %p')
time=pd.DataFrame(creating_a_timerange)
time.head()



```

```python
ambient=np.random.rand(380,1)

len(ambient)

ambient = 23 + (ambient * (24 - 23))
print(len(ambient))



ambient1=np.round(ambient,2)



print(ambient1)
```

```python
import matplotlib.pyplot as plt

df["Temperature"] = pd.to_numeric(df["Temperature"], errors='coerce')  

time_data = time[0][::10]
temperature_data = df["Temperature"][::10]

plt.figure(figsize=(10, 6))

plt.plot(time_data, temperature_data, color='r', linewidth=2, marker='*', markersize=8)
plt.bar(time_data, temperature_data, color='orange')
min_temp = df["Temperature"].min()
max_temp = df["Temperature"].max()



plt.ylim(min_temp-.5, max_temp+.5)
plt.tick_params(axis='y', labelrotation=90)
plt.xticks(rotation=90)
plt.grid()
plt.xlabel("Time",color='black')
plt.ylabel("Temperature")
plt.title("Temperature vs time domain representation",fontsize=12)

plt.show()

```

```python
fig,ax1=plt.subplots()

time_data = time[0][::20]
temperature_data = df["Temperature"][::20]
ambient_temperature_data=ambient1[::20]

ax1.plot(time_data, temperature_data, color='r', linewidth=2, marker='*', markersize=8)
plt.xticks(rotation='vertical')
ax1.grid(True)
ax1.set_xlabel("Time Domain")
ax1.set_ylabel("Colling room Temperature[\u00B0C]")
ax1.set_ylim([14, 30])



ax2=ax1.twinx()
ax2.plot(time_data,ambient_temperature_data,color='green', linewidth=2, marker='o', markersize=8)
ax2.set_ylim([14,30])
ax2.set_ylabel("Ambient Temperature[\u00B0C]")
ax2.grid(True)
```
